```python
from pandas import json_normalize, to_datetime
import json
```

```python
with open('metrics.txt', 'r') as file:
    lines = file.readlines()

# Parse each line as a JSON object
data = [json.loads(line) for line in lines]
```

```python
df = json_normalize(data)
df
```

```python
connecting = df[df["metric"] == "ws_connecting"][df["type"] == "Point"]
# connecting["time"] = to_datetime(connecting["data.time"], unit='s', errors='coerce')
```

```python
# graph a line chart of data.time of connecting
import matplotlib.pyplot as plt

# smooth the line

plt.plot(connecting["data.time"], connecting["data.value"])

plt.show()

```
